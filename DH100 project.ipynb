### Importing libraries ###

```python
import os
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline 
```

### Loading data into Pandas ###

```python
report1 = pd.read_csv('./kaggle - world happiness report/world-happiness-report.csv')
report2 = pd.read_csv('./kaggle - world happiness report/world-happiness-report-2021.csv')

# An alternative of reading the csv file into a dataframe
# happy_all = pd.read_csv('kaggle - world happiness report/world-happiness-report.csv')
```

### Renaming Columns ###

```python
report1.rename(columns={'Life Ladder': 'Ladder score', 'Healthy life expectancy at birth': 'Healthy life expectancy'}, inplace=True)

# reference: https://stackoverflow.com/questions/11346283/renaming-columns-in-pandas
```

### World Happiness Report 2021 with the year column ###

```python
report2.insert(1,'year', 2021)
report2

#reference: https://stackoverflow.com/questions/18674064/how-do-i-insert-a-column-at-a-specific-column-index-in-pandas
```

### World Happiness Report 2005 to 2020 with Regional indicator ###

```python
country_region = report2[['Country name', 'Regional indicator']]
report1_region = report1.merge(country_region) #this takes out the missing country names from report_2021
report1_region
```

### Cleaning Report 1 and Report 2 
Making report 1 and report 2 have the same column names and sorted the same way. Also, keeping only the country's that are in both dataframes.

```python
# 2005-2020: drop columns that do not match report2
report1 = report1_region.drop(['Positive affect', 'Negative affect'], axis = 1)

# 2021: drop columns that do not match report1 and sort same way as report 1
report2 = report2.rename(columns = {'Logged GDP per capita' : 'Log GDP per capita'})
report2 = report2[report1.columns]
report2.sort_values('Country name', ascending = True, inplace = True)
report2.reset_index(drop = True, inplace = True)
report2

# remove countries that are in both report1 and report2
s1 = report1['Country name']
s2 = report2['Country name']
s1 = s1.drop_duplicates()
s1.reset_index(drop=True, inplace=True)
pd.concat([s1, s2]).drop_duplicates(keep=False) # end result shows that both dataframes have the same countries, do not need removal
```

### Correlation between Report 1 and Report 2
It seems that there is a positive correlation between these two variables. <br>
People have a longer life expectancy if they are happier in life.

```python
ax = report1.plot.scatter(x = 'Healthy life expectancy', y = 'Ladder score', color = 'Orange', label = '2005 - 2020')
report2.plot.scatter(x = 'Healthy life expectancy', y = 'Ladder score', color = 'DarkGreen', ax = ax, label = '2021')
# source reference:
# https://stackoverflow.com/questions/13872533/plot-different-dataframes-in-the-same-figure
# https://pandas.pydata.org/pandas-docs/stable/user_guide/visualization.html#scatter-plot
```

### Experimental Cell

```python
#get dataframe columns to numpy arrays
ladderScore1 = report1['Ladder score'].to_numpy() 
ladderScore2 = report2['Ladder score'].to_numpy() 

healthyLife1 = report1['Healthy life expectancy'].to_numpy()
healthyLife2 = report2['Healthy life expectancy'].to_numpy()
#use column arrays to find r (correlation)

r1 = np.corrcoef(healthyLife1, ladderScore1)
r2 = np.corrcoef(healthyLife2, ladderScore2)

r1, r2

##compare correlation between 2005-2020 and 2021
```

### Correlation between Report 1 mean and Report 2
The blue dots show the mean of the values throughout the years 2005 - 2020, while the green dots show only the year 2021. Through this data, it demonstrates that although 2021 had lower ladder scores than the mean of the 2005 to 2020 years, it also had the highest ladder scores as well as healthy life expectancy. According to the average, the happiness levels before the pandemic and after shows to be false.

```python
report1_noYear = report1_region.drop(['year'], axis = 1)
report1_meanByCountry = report1_noYear.groupby(['Country name']).mean()
ax = report1_meanByCountry.plot.scatter(x = 'Healthy life expectancy', y = 'Ladder score', color = 'Orange', label = '2005 - 2020')
report2.plot.scatter(x = 'Healthy life expectancy', y = 'Ladder score', color = 'DarkGreen', ax = ax, label = '2021')

report1_meanByCountry.reset_index(inplace = True)
```

```python
#get dataframe columns to numpy arrays
ladderScoreA = report1_meanByCountry['Ladder score'].to_numpy() 
ladderScoreB = report2['Ladder score'].to_numpy() 

healthyLifeA = report1_meanByCountry['Healthy life expectancy'].to_numpy()
healthyLifeB = report2['Healthy life expectancy'].to_numpy()
#use column arrays to find r (correlation)

rA = np.corrcoef(healthyLifeA, ladderScoreA)
rB = np.corrcoef(healthyLifeB, ladderScoreB)

r1, r2

##compare correlation between 2005-2020 mean and 2021
```

### Ladder score according to Regional indicator (2005 - 2020)
The blue vertical bar chart shows the data of 2005 to 2020 mean values of the ladder scores in increasing order corresponding to specific regions.

```python
# grouping by region for dataset 2005 - 2020
report1_meanByRegion = report1_noYear.groupby(['Regional indicator']).mean()
report1_meanByRegion.reset_index(inplace = True)

sns.factorplot(x='Ladder score', y='Regional indicator', kind='bar', data=report1_meanByRegion)
```

### Ladder score according to Regional indicator (2021)
The green vertical bar chart shows the data of 2021 values of the ladder scores in increasing order corresponding to specific regions.

```python
#grouping by region for dataset 2021
report2_meanByRegion = report2.groupby(['Regional indicator']).mean()
report2_meanByRegion.reset_index(inplace = True)

sns.factorplot(x='Ladder score', y='Regional indicator', kind='bar', data=report2_meanByRegion)
```

### Comparison of Ladder score between 2005-2020 and 2021 (Bar chart version)
Report 1 is data from 2005 to 2020 and report 2 is from 2021. This is a comparison between the two datasets of happiness levels by regional indicator. From the plot below, it shows that people were more satisfied with their lives in 2021 than previous years as report 2 demonstrates to be higher in most regions. The other regions are either lagging slightly behind or no major difference. 
<br><br>
However, it is important to notice that the ladder score of this chart shows all regions are below the score 8. Meaning, happiness levels may have increased throughout the regions in 2021, but there is causing factors for it to be lower than 8.
#### Note: have to filter out the country names in the report from 2005 to 2020 that does not match 2021 data ####

```python
r1_region_ladder = report1_meanByRegion[['Regional indicator', 'Ladder score']]
r2_region_ladder = report2_meanByRegion[['Regional indicator', 'Ladder score']]
r1_r2_meanByRegion = pd.concat([r1_region_ladder, r2_region_ladder])
r1_r2_meanByRegion['report'] = (len(r1_region_ladder)*(1,) + len(r2_region_ladder)*(2,))
r1_r2_meanByRegion.reset_index(inplace = True)
r1_r2_meanByRegion 
sns.factorplot(x='Ladder score', y='Regional indicator', hue='report', kind='bar', data=r1_r2_meanByRegion)
#reference: https://stackoverflow.com/questions/42532319/grouped-bar-chart-from-two-pandas-data-frames
```

### Ladder Score Histogram (2005 - 2021)
The histogram shows the KDE density and different percentiles of the data.

```python
# https://towardsdatascience.com/take-your-histograms-to-the-next-level-using-matplotlib-5f093ad7b9d3
ladScore1 = report1_meanByCountry['Ladder score']

fig1, ax1 = plt.subplots(figsize = (6, 4))
ladScore1.plot(kind = 'hist', density = True, alpha = 0.65, bins = 15)
ladScore1.plot(kind = 'kde')

# x
ax1.set_xlabel('Ladder score')
# y
ax1.set_ylim(0, 0.45)

#style
# ladScore.plot(kind = 'hist', density = True, alpha = 0.65, bins = 15)
ax1.grid(False)
ax1.set_title('Ladder Score Around The World (2005 - 2020)', size = 17, pad = 10)
ax1.set_ylabel("")

#calculate percentiles
quant1_5, quant1_25, quant1_50, quant1_75, quant1_95 = ladScore1.quantile(0.05), ladScore1.quantile(0.25), ladScore1.quantile(0.50), ladScore1.quantile(0.75), ladScore1.quantile(0.95)
# [quantile, opacity, length]
quants1 = [[quant1_5, 0.6, 0.16], [quant1_25, 0.8, 0.26], [quant1_50, 1.0, 0.36], [quant1_75, 0.8, 0.46], [quant1_95, 0.6, 0.56]]
#plot lines with loop
for i in quants1:
    ax1.axvline(i[0], alpha = i[1], ymax = i[2], linestyle = ':')

# Text description of percentiles
ax1.text(quant1_5-.2, 0.08, "5th", size = 11, alpha = 1)
ax1.text(quant1_25-.25, 0.12, "25th", size = 11, alpha = 1)
ax1.text(quant1_50-.30, 0.17, "50th", size = 11, alpha = 1)
ax1.text(quant1_75-.50, 0.21, "75th", size = 11, alpha = 1)
ax1.text(quant1_95-.60, 0.255, "95th Percentile", size = 11, alpha = 1)
```
